```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load models and tokenizers
llama_model = AutoModelForCausalLM.from_pretrained('meta-llama/Meta-Llama-3.1-8B',token="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")


gemini_model = AutoModelForCausalLM.from_pretrained('path/to/gemini-model')
gemini_tokenizer = AutoTokenizer.from_pretrained('path/to/gemini-tokenizer')

def evaluate_model(model, tokenizer, prompt):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Example prompts
java_prompt = "can u give me a java methode that realize a bank account "
Create a Java function that determines if a given number is prime.
# Get responses from both models
llama_response = evaluate_model(llama_model, llama_tokenizer, java_prompt)
gemini_response = evaluate_model(gemini_model, gemini_tokenizer, java_prompt)

print("LLaMA Response:", llama_response)

```

```python
from deepeval import assert_test
from deepeval.test_case import LLMTestCase
from deepeval.metrics import AnswerRelevancyMetric

def test_answer_relevancy():
    answer_relevancy_metric = AnswerRelevancyMetric(threshold=0.5)
    test_case = LLMTestCase(
        input="What if these shoes don't fit?",
        # Replace this with the actual output of your LLM application
        actual_output="We offer a 30-day full refund at no extra cost."
    )
    assert_test(test_case, [answer_relevancy_metric])
```
